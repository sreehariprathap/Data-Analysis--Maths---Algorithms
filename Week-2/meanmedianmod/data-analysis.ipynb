### Summary
This dataset includes details of 5,659 Indian movies from 1931 to 2021 on IMDB. It covers movie name, release year, duration, genre, rating, votes, director, and top three actors. It can be used to analyze top movies each year, actor combinations affecting ratings, genre preferences, and duration's impact on ratings.

### Data 
- **Sample**: 5,659 Indian movies listed on IMDB.
- **Population**: All Indian movies released from 1931 to 2021.


```python
import pandas as pd
from scipy import stats
from tabulate import tabulate

# Read the CSV file into a DataFrame
df = pd.read_csv('./movies_data.csv')

# Display the first few rows of the DataFrame in a formatted table
print("First few rows of the dataset:")
print(tabulate(df.head(), headers='keys', tablefmt='grid'))

# Calculate Mean, Median, and Mode for 'Rating'
rating_mean = df['Rating'].mean()
rating_median = df['Rating'].median()
rating_mode = df['Rating'].mode()[0]

# Print the results
print("\nStatistics for Ratings:")
print(f"Mean: {rating_mean:.2f}")
print(f"Median: {rating_median:.2f}")
print(f"Mode: {rating_mode}")
```

### Descriptive Statistics

- **Mean**: The average value, such as the average movie rating, duration, or votes, provides an overall trend.
- **Median**: The middle value when data is sorted, which is less affected by outliers, helps understand the central tendency of movie ratings or durations.
- **Mode**: The most frequently occurring value, such as the most common genre or director, indicates popular trends or preferences.

### Implications for Use Case

- **Mean**: Shows overall trends and averages, useful for general insights into movie ratings or durations.
- **Median**: Provides a robust central value, especially when analyzing skewed data distributions in ratings or movie lengths.
- **Mode**: Identifies the most common genres or actors, helping understand prevalent trends and popular choices in Indian cinema.
